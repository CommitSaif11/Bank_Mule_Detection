```python
import pandas as pd

df = pd.read_csv("../data/DataSet.csv")
```

```python
print("=" * 60)
print("Shape of dataset:", df.shape)
```

```python
print("=" * 60)
print("Target variable (F3924) value counts and percentages:")
print(df["F3924"].value_counts())
print(df["F3924"].value_counts(normalize=True) * 100)
```

```python
null_counts = df.isnull().sum()
n_rows = len(df)
fully_null = (null_counts == n_rows).sum()
gt80 = (null_counts / n_rows > 0.8).sum()
gt50 = (null_counts / n_rows > 0.5).sum()

print("=" * 60)
print("Number of fully null columns:", fully_null)
print("Number of columns with >80% missing values:", gt80)
print("Number of columns with >50% missing values:", gt50)
```

```python
print("=" * 60)
print("Data types count:")
print(df.dtypes.value_counts())
```

```python
print("=" * 60)
overall_missing = df.isnull().sum().sum() / (df.shape[0] * df.shape[1]) * 100
print(f"Overall missing value percentage: {overall_missing:.4f}%")
```

```python
cols = ["F115", "F321", "F527", "F531", "F670", "F1692", "F2082", "F2122",
        "F2582", "F2678", "F2737", "F2956", "F3043", "F3836", "F3887",
        "F3889", "F3891", "F3894"]
```

```python
print("=" * 60)
print("Stats for specific columns:")
for col in cols:
    if col not in df.columns:
        print(f"\n{col}: NOT FOUND IN DATASET")
        continue
    s = df[col]
    print(f"\n--- {col} (dtype={s.dtype}) ---")
    print(f"  null count: {s.isnull().sum()}")
    print(f"  nunique: {s.nunique()}")
    if pd.api.types.is_numeric_dtype(s):
        print(f"  mean: {s.mean()}")
        print(f"  std: {s.std()}")
    else:
        print("  mean: N/A (non-numeric)")
        print("  std: N/A (non-numeric)")
```

```python
print("=" * 60)
print("Mean by fraud class (F3924) for numeric columns:")
for col in cols:
    if col not in df.columns:
        continue
    if pd.api.types.is_numeric_dtype(df[col]):
        means = df.groupby("F3924")[col].mean()
        print(f"\n--- {col} ---")
        print(means)
```

```python
print("=" * 60)
print("F3889 value counts by fraud class (F3924):")
print(df.groupby("F3924")["F3889"].value_counts())
```

```python
print("=" * 60)
print("F3891 value counts by fraud class (F3924):")
print(df.groupby("F3924")["F3891"].value_counts())
```
